```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_from_file(path, starting_at = 80):
    json_txt = ""
    with open(path) as file:
        json_txt = "".join(file.readlines())
    data = json.loads(json_txt)
    # print(data)
    plt.rcParams["figure.figsize"] = (#35
    20,10)
    power = [key for key in data]
    berc = [val['BERC'] for key, val in data.items()]
    ser = [val['SER'] for key, val in data.items()]
    iters = [val['iters'] for key, val in data.items()]
    f, ax = plt.subplots()
    ax.plot(power[starting_at:], iters[starting_at:])
    ax.plot(power[starting_at:], berc[starting_at:])
    ax.plot(power[starting_at:], ser[starting_at:])
    plt.legend(('Number of iterations', 'BER', 'FER'))
    plt.ylabel("Count")
    plt.xlabel("Gaussian channel noise power")
    


    xmin, xmax = ax.get_xlim()
    ax.set_xticks(np.round(np.linspace(xmin, xmax, 10), 2))

    plt.show()
```

```python
plot_from_file("build/result.json")
```

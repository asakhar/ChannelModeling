```python
import json
import matplotlib.pyplot as plt
```

```python
def plot_from_file(path, starting_at = 80):
    json_txt = ""
    with open(path) as file:
        json_txt = "".join(file.readlines())
    data = json.loads(json_txt)
    # print(data)
    plt.rcParams["figure.figsize"] = (35,10)
    power = [key for key in data]
    berc = [val['BERC'] for key, val in data.items()]
    ser = [val['SER'] for key, val in data.items()]
    iters = [val['iters'] for key, val in data.items()]
    plt.plot(power[starting_at:], iters[starting_at:])
    plt.plot(power[starting_at:], berc[starting_at:])
    plt.plot(power[starting_at:], ser[starting_at:])

    plt.show()
```

```python
plot_from_file("build/result.json")
```
